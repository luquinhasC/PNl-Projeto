```julia
using ForwardDiff
using LinearAlgebra
using Plots
```

```julia
## Em Km ##
u_1 = [ 3.718 ,-1.847] ##Poli
u_2 = [ 2.177 , 2.289] ##Agrarias
u_3 = [ 0.861 , 0.728] ##Prédio Histórico
u_4 = [ 0.616 ,-0.405] ##Rebouças
u_5 = [-1.837 , -1.997] ##Puc
#U = [u_1,u_2,u_3,u_4,u_5]
U= [u_1,u_2,u_3]
λ_1 = 4
λ_2 = λ_3 = 2
λ_4 = λ_5 = 1
λ = [λ_1,λ_2,λ_3,λ_4,λ_5]
function f(x_1,x_2)
    y = 0
    for i in range(1, stop = size(U,1))
       y = y + λ[i]*sqrt((x_1 - U[i][1])^2 + (x_2 - U[i][2])^2 ) 
    end
   return y
end
```

```julia
## Em Km ##
u_1 = [ 3.718 ,-1.847] ##Poli
u_2 = [ 2.177 , 2.289] ##Agrarias
u_3 = [ 0.861 , 0.728] ##Prédio Histórico
u_4 = [ 0.616 ,-0.405] ##Rebouças
u_5 = [-1.837 , -1.997] ##Puc
U = [u_1,u_2,u_3,u_4,u_5]
#U= [u_1,u_2,u_3]
λ_1 = 4
λ_2 = λ_3 = 2
λ_4 = λ_5 = 1
λ = [λ_1,λ_2,λ_3,λ_4,λ_5]
function f(x)
    x_1 = x[1]
    x_2 = x[2]
    y = 0
    for i in range(1, stop = size(U,1))
       y = y + λ[i]*sqrt((x_1 - U[i][1])^2 + (x_2 - U[i][2])^2 ) 
    end
   return y
end
# É necessário rodar as duas celulas por que a biblioteca ForwardDiff utiliza um método
# e a pyplot outra
```

```julia
P_i = [-2,-2]
∇f = ForwardDiff.gradient(f,P_i)
```

```julia
pyplot()
```

```julia
grid = range(-5,5 ,length = 100) # Plotagem da função, mudar a função acima caso queira plotar outras funções
surface(grid,grid,f ,camera = (40,40),
 title = "Distância aos campus" ,
 xlabel = "x" ,
 ylabel = "y",
 )
```

```julia
g = range(-5,5 ,length = 100) #Analogo À acima mas para curvas de nivel
contour(g,g,f)
scatter!([P_i[1]],[P_i[2]], c = :blue)
plot!([P_i[1],P_i[1] - 1*∇f[1] ],[P_i[2],P_i[2]- 1*∇f[2] ] , c = :green, arrow = :arrow)
```
